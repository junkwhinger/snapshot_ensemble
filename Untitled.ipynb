```python
tt = 0
TeNext = Te ## initial value: 50
batchsize = 4 ## 128.
```

```python
## 6. training loop
```

```python
Estart
```

```python
labelcnt = 40
```

```python
lrs = []
clrs = []
tts = []
dts = []
curTs = []
## for every epoch...
for epoch in range(num_epochs):
    ## shuffle training data
    ## do a full pass over the training data
    
    for batch in range(batchsize):
        print("E{:03}-B{:1}".format(epoch, batch))
        
        ## time to start adjust learning rate
        if (epoch + 1 >= Estart):
            ## Te: repeat every
            dt = 2.0 * math.pi / float(2.0 * Te)
            print(Te, dt, tt)
            tt = tt + float(dt) / (labelcnt / batchsize)
            if tt >= math.pi:
                tt = tt - math.pi
                print("new tt: {}".format(tt))
            tts.append(tt)
            dts.append(dt)
            
            curT = t0 + tt
            curTs.append(curT)
            
            print(">>", t0, tt, curT)
#     new_lr = 0 + 1/2 * (lr - 0) * (1 + np.cos((curT / Te) * np.pi))
            new_lr = lr * (1.0 + math.sin(curT)) / 2.0
    t_cur = epoch % Te
    cnew_lr = 0 + 1/2 * (lr - 0) * (1 + np.cos((t_cur / Te) * np.pi))
    clrs.append(cnew_lr)
    lrs.append(new_lr)
#             print("start adjust learning rate")
            
    if (epoch + 1 == TeNext):
        tt = 0
        Te = Te * multiFactor
        TeNext = TeNext + Te
        
#     print(epoch)
```

```python
import numpy as np
```

```python
def calc_lr(since_lrs, rs_every, lr_max, lr_min):
    new_lr = lr_min + 1/2 * (lr_max - lr_min) * (1 + np.cos(since_lrs / rs_every * np.pi))
    return new_lr
```

```python
calc_lr(5, 5)
```

```python
lrs = []

total_epochs = 200
Te = 10

TeNext = Te
multFactor = 2
e_cum = 0
for i in range(0, total_epochs + 1):  
    
    new_lr = calc_lr(e_cum, Te, lr_max=5e-2, lr_min=5e-4)
    
    print("epoch: {}, cumulative_run: {}, new_lr: {}".format(i, e_cum, new_lr))

    if i > 0:
        if i % TeNext == 0:
            e_cum = 0
            print("RESET--", i, TeNext)
        
    if i == TeNext:
        Te = Te * multFactor
        TeNext = TeNext + Te
        print(i, TeNext)
    
    lrs.append(new_lr)
    e_cum += 1
    
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(lrs)
plt.show()
```
